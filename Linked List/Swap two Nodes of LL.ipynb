### Swap two Nodes of LL

You have been given a singly linked list of integers along with two integers, 'i,' and 'j.' Swap the nodes that are present at the 'i-th' and 'j-th' positions.

```python
class Node(object):
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList(object):
    
    def __init__(self):
        self.head = None
        
    def createLL(self,arr):
        
        for item in arr:
            new_node = Node(item)
            if self.head == None:
                self.head = new_node
                
            else:
                current_node = self.head
                while current_node.next is not None:
                    current_node = current_node.next
                    
                current_node.next = new_node
    
    def swap(self,i,j):
        
        if self.head is None:
            return None
        if i == j:
            return self.head
        if i > j:
            i,j = j,i
        k =0
        first = self.head
        second = self.head
        if j > i and i !=0:
            k = j-i
        
        elif i == 0:
            k = j-1
        
        current = self.head
        imoves,jmoves = 0,0
       
                 
        
        while current is not None and imoves < i-1 and i != 0:
            first = first.next
            second = second.next
            current = current.next
            imoves = imoves+1
            
        while current is not None and jmoves < k:
            second = second.next
            current = current.next
            jmoves = jmoves+1
            
   
        next_first = first.next
        next_second = second.next
       
        if i ==0 and abs(i-j) == 1:
            next_first = first.next
            next_next_first = first.next.next
            self.head = next_first
            next_first.next = first
            first.next = next_next_first
            return
            
        if i ==0:
            self.head = next_second
            next_second.next = next_first
            second.next = first
            first.next = None
            return
            
            
        next_next_first = first.next.next
        next_next_second = second.next.next

        if j-i == 1:
      
            first.next = next_second
            next_second.next = next_first
            next_first.next = next_next_second
            return
        
        first.next = next_second
        next_second.next = next_next_first
        
        second.next = next_first
        next_first.next = next_next_second
      

    def printLL(self):
        
        temp = self.head
        while temp is not None:
            print(temp.data,end="->")
            temp = temp.next
        print(None,end="")
```

```python
a = [3, 4, 5, 2, 6, 1, 9]
linkedlist = LinkedList()
linkedlist.createLL(a)
linkedlist.printLL()
print()
linkedlist.swap(0,1)
linkedlist.printLL()
```
