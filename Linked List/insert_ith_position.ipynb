```python
class Node():
    
    def __init__(self,data):
        self.data = data
        self.next = None


class LinkedList():
    
    def __init__(self):
        self.head = None
        self.length = 0
        
    def createLL(self,arr):
        
        for data in arr:

            new_node = Node(data)
            if self.head == None:
                self.head = new_node
                current_node = new_node
            
            else:
                current_node.next = new_node
                current_node = new_node
                
    def insert_ipos(self, i, data):
        
        if i > self.lengthLL() or i < 0:
            return 
        
        if i == 0: #first position, need to change head
            
            new_node = Node(data)
            node = self.head
            self.head = new_node
            new_node.next = node
            
            return
        
        temp = self.head
        for _ in range(i-1):
            temp = temp.next
        
        new_node = Node(data)
        next_node = temp.next
        temp.next = new_node
        new_node.next = next_node
        
    def lengthLL(self):
        temp = self.head
        
        while temp is not None:
            self.length = self.length +1
            temp = temp.next
        
        return self.length
        
    def printLL(self):
        temp = self.head
        
        while temp is not None:
            print(temp.data, end = "->")
            temp = temp.next
        print("None")

if __name__ == "__main__":
    
    user_input = list(map(int,input().split()))
    
    linkedlist = LinkedList()
    linkedlist.createLL(user_input)
    linkedlist.insert_ipos(2,6)
    linkedlist.printLL()
    linkedlist.insert_ipos(0,9)
    linkedlist.printLL()
    linkedlist.insert_ipos(7,10)
    linkedlist.printLL()
```

### There is another approach where we male use of previous and current pointers
